```python
import tkinter as tk
from tkinter import messagebox

# ================== APPLICATION CLASS ================== #
class StudentResultAnalyzer:

    def __init__(self, root):
        self.root = root
        self.root.title("Student Result Analyzer System")
        self.root.geometry("840x720")
        self.root.configure(bg="#eef2f7")
        self.root.resizable(False, False)

        self.subjects = ["Computer", "Physics", "Chemistry", "English", "Maths"]
        self.entries = {}

        self.create_header()
        self.create_input_section()
        self.create_buttons()
        self.create_result_section()

    # ================== HEADER ================== #
    def create_header(self):
        header = tk.Frame(self.root, bg="#1f4e79", height=80)
        header.pack(fill="x")

        title = tk.Label(header,
                         text="🎓 Student Result Analyzer",
                         bg="#1f4e79",
                         fg="white",
                         font=("Segoe UI", 20, "bold"))
        title.pack(pady=20)

    # ================== INPUT SECTION ================== #
    def create_input_section(self):
        input_frame = tk.Frame(self.root, bg="white", bd=0)
        input_frame.pack(pady=20, padx=40, fill="x")

        tk.Label(input_frame,
                 text="Enter Subject Marks (0 - 100)",
                 font=("Segoe UI", 14, "bold"),
                 bg="white",
                 fg="#333").pack(pady=10)

        form_frame = tk.Frame(input_frame, bg="white")
        form_frame.pack(pady=10)

        for i, subject in enumerate(self.subjects):
            tk.Label(form_frame,
                     text=subject,
                     font=("Segoe UI", 11),
                     bg="white").grid(row=i, column=0, pady=8, sticky="w")

            entry = tk.Entry(form_frame,
                             font=("Segoe UI", 11),
                             width=15,
                             relief="solid")
            entry.grid(row=i, column=1, pady=8, padx=10)
            self.entries[subject] = entry

    # ================== BUTTON SECTION ================== #
    def create_buttons(self):
        btn_frame = tk.Frame(self.root, bg="#eef2f7")
        btn_frame.pack(pady=10)

        calculate_btn = tk.Button(btn_frame,
                                   text="Calculate Result",
                                   font=("Segoe UI", 12, "bold"),
                                   bg="#1abc9c",
                                   fg="white",
                                   width=18,
                                   command=self.calculate_result)
        calculate_btn.grid(row=0, column=0, padx=10)

        reset_btn = tk.Button(btn_frame,
                              text="Reset",
                              font=("Segoe UI", 12, "bold"),
                              bg="#e74c3c",
                              fg="white",
                              width=10,
                              command=self.reset_fields)
        reset_btn.grid(row=0, column=1, padx=10)

    # ================== RESULT SECTION ================== #
    def create_result_section(self):
        self.result_frame = tk.Frame(self.root,
                                     bg="white",
                                     bd=2,
                                     relief="groove")
        self.result_frame.pack(pady=20, padx=40, fill="both", expand=True)

        self.result_label = tk.Label(self.result_frame,
                                     text="Result will appear here",
                                     font=("Segoe UI", 12),
                                     bg="white",
                                     justify="left")
        self.result_label.pack(pady=20)

    # ================== LOGIC ================== #
    def calculate_result(self):
        marks = {}
        is_pass = True

        try:
            for subject in self.subjects:
                score = int(self.entries[subject].get())

                if score < 0 or score > 100:
                    messagebox.showerror("Invalid Input",
                                         f"{subject} marks must be between 0 and 100.")
                    return

                marks[subject] = score

                if score < 40:
                    is_pass = False

            total = sum(marks.values())
            percentage = total / len(self.subjects)

            if percentage >= 90:
                grade = "A"
            elif percentage >= 75:
                grade = "B"
            elif percentage >= 60:
                grade = "C"
            elif percentage >= 40:
                grade = "D"
            else:
                grade = "F"

            status = "PASS" if is_pass and grade != "F" else "FAIL"

            result_text = (
                "------------- RESULT SUMMARY -------------\n\n"
                f"Total Marks   : {total}\n"
                f"Percentage    : {percentage:.2f}%\n"
                f"Grade         : {grade}\n"
                f"Status        : {status}\n"
                "\n------------------------------------------"
            )

            self.result_label.config(text=result_text)

        except ValueError:
            messagebox.showerror("Error", "Please enter valid numeric marks.")

    def reset_fields(self):
        for entry in self.entries.values():
            entry.delete(0, tk.END)
        self.result_label.config(text="Result will appear here")


# ================== ROOT WINDOW ================== #
if __name__ == "__main__":
    root = tk.Tk()
    app = StudentResultAnalyzer(root)
    root.mainloop()
```
